# Basic images load, plot, resize, blur, etc...

```python
import numpy as np
import os
from scipy.misc import imread, imresize
import skimage.io
import skimage.transform
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#Print current folder
cwd = os.getcwd()
print('Current folder is: ', cwd)

#Useful function
def print_typeshape(img):
    print('Type is: ', type(img))
    print('Shape is: ', img.shape)
```

# Load & Plot image

```python
#Load
img = imread('images/image1.jpg')
print_typeshape(img)

#Plot
plt.figure(0)
plt.imshow(img)
plt.title('Original Image with imread')
```

```python
#Load images with float type
img2 = imread('images/image1.jpg').astype(np.float32)
print_typeshape(img2)

#Plot
plt.figure(0)
plt.imshow(img2)
plt.title('Original Image with imread.astype(np.float32)')
```

```python
#Load images with float type
img3 = imread('images/image1.jpg').astype(np.float32)
print_typeshape(img3)

#Plot
plt.figure(0)
plt.imshow(img3/255.)
plt.title('Original Image with imread.astype(np.float32)/255.')
```

# Resize

```python
img_small = imresize(img, [100, 100, 3])
print_typeshape(img_small)

#Plot
plt.figure(0)
plt.imshow(img_small)
plt.title('Resized Image')
```

# Grayscale

```python
#convert to grayscale function
def rgb2gray(rgb_img):
    if len(rgb_img.shape) is 3:
        return np.dot(rgb_img[...,:3], [0.299, 0.587, 0.114])
    else:
        print('Current Image if Gray')
        return rgb_img

img_gray = rgb2gray(img)

print_typeshape(img_gray)

#Plot
plt.figure(0)
plt.imshow(img_gray, cmap=plt.get_cmap('gray'))
plt.title('Gray Scale Image')
plt.colorbar()
```

# Reshape

```python
img_small_gray = rgb2gray(img_small)
#Convert image_small_gray to Vector
img_vec = np.reshape(img_small_gray, newshape=(1,-1))
print_typeshape(img_vec)

#Convert image_small_gray back to Matrix
img_matrix = np.reshape(img_small_gray, newshape=(100, 100))
print_typeshape(img_matrix)
```

# Load images from folder

```python
#Path to the folder want to load images
path = 'C:/Users/StrikeWade/Desktop/hinh/hinh/'
valid_file_type = ['.jpg', '.gif', '.png', '.jpeg']

#Print number of images in folder
print('%d images in %s' % (len(os.listdir(path)), path))

#Append Images and their names to lists
imgs = []
names = []

for f in os.listdir(path):
    #For all files
    ext = os.path.splitext(f)[1]
    
    #Check types
    if ext.lower() not in valid_file_type:
        continue
    fullpath = os.path.join(path, f)
    print(fullpath)
    imgs.append(imread(fullpath))
    names.append(os.path.splitext(f)[0] + os.path.splitext(f)[1])

print('%d images loads' % (len(imgs)))
```

```python
#Check images & Plot 
num_imgs = len(imgs)
random_idx = np.random.randint(num_imgs, size=5)
print('Types of imgs: ', type(imgs))
print('Length of imgs: ', num_imgs)

for current_img, current_name, i in zip([imgs[j] for j in random_idx], [names[j] for j in random_idx], range(len(random_idx))):
    print('[Image %d] Type of current_img: %s' % (i+1, type(current_img)))
    print('     Name of the current_img is: ', current_name)
    print('     Shape of the current_img is: ', current_img.shape)
    
    plt.figure(i)
    plt.imshow(current_img)
    plt.title('[Image ' + str(i+1) + ']')
```
